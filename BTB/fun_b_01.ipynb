```python
class Areaofshapes():
    
    @staticmethod
    def Area(i):
        if i==1:
            #square
            a=float(input("Enter value of the side"))
            return f"{a**2:0.3f}"
        elif i==2:
            #rectangle
            l=float(input("Enter value of the l-side"))
            b=float(input("Enter value of the b-side"))
            return f"{(l*b):0.3f}"
        elif i==3:
            #tringle
            a=float(input("Enter value of the 1st-side"))
            b=float(input("Enter value of the 2nd-side"))
            c=float(input("Enter value of the 3rd-side"))
            s = (a+b+c)/2
            area = math.sqrt(s*(s-a)*(s-b)*(s-c))
            return f"Area of Triangle = {area:0.3f}"
    
        
    @staticmethod
    def perimeter(i):
        if i==1:
            #square
            peri=float(input("Enter the value fo the side"))
            return f"{4*peri:0.3f}"
        elif i==2:
            #rectangle
            l=float(input("Enter value of the l-side"))
            b=float(input("Enter value of the b-side"))
            return f"{2*(l+b):0.3f}"
        elif i==3:
             # triangle
            a = float(input("Enter value of the 1st side: "))
            b = float(input("Enter value of the 2nd side: "))
            c = float(input("Enter value of the 3rd side: "))
            return f"Perimeter of Triangle = {a+b+c:0.3f}"
        
obj=Areaofshapes()
print("-------calciiiii-------")
while True:
    print("1.square")
    print("2.rectangle")
    print("3.triangle")
    print("4.Exit")
    choice=int(input(">"))
    if choice in [1,2,3]:
        print("1.area")
        print("2.perimeter")
        choice_2=int(input(">"))
        if choice_2==1:
            print(obj.Area(choice))
        elif choice_2==2:
            print(obj.perimeter(choice))
        else:
            print("invalid-option")    
    else:
        print("bye")
        break
```
